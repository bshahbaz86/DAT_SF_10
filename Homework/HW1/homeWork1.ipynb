```python
#Load JSON to python environment the import login file
import json    
json_data = open('ga_hw_logins.json')
data = json.load(json_data)
#Dropping the "u'", line breaks and white spaces
data = json.dumps(data, indent=1)
data.replace('\n','').replace(' ','')

# print data #preview data
```

```python
from datetime import datetime

#load string as json
dt = json.loads(data)

#create a list to be populated in the for loop
dateTimeList = []

for items in dt[1:-1]:
    dateTime = datetime.strptime(items, '%Y-%m-%d %H:%M:%S') #convert string to datetime object
    dateTimeList.append(dateTime)
print dateTimeList

```
