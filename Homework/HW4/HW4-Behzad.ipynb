```python
# imports
from sklearn.datasets import fetch_olivetti_faces
from sklearn.cluster import MiniBatchKMeans
from sklearn.decomposition import PCA
from sklearn.decomposition import RandomizedPCA
from numpy.random import RandomState
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import pairwise_distances
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
import pandas as pd
%matplotlib inline
```

```python
# Load faces data
rng = RandomState(0)
dataset = fetch_olivetti_faces(shuffle=True, random_state=rng)
faces = dataset.data
n_samples, n_features = faces.shape

# global centering
faces_centered = faces - faces.mean(axis=0)

# local centering
faces_centered -= faces_centered.mean(axis=1).reshape(n_samples, -1)
print("Dataset consists of %d faces" % n_samples)
```

## 1. Implement PCA on the dataset, using the sklearn. What do the principal components tell? (is scaling necessary?)

```python
X = faces
y = faces.shape
target_names = faces.shape

# create the model and fit the data
pca = PCA(n_components=400)
X_r = pca.fit(X).transform(X)
ratios = pca.explained_variance_ratio_
print"First two components: %f" % (100*ratios[0:2].sum())
print"First five components: %f" % (100*ratios[0:5].sum())
print"First ten components: %f" % (100*ratios[0:10].sum())
print"First twenty components: %f" % (100*ratios[0:20].sum())
```

### Frist twenty components account for 76% of variance

```python
comp_id = xrange(1,401)             # id number of component
fig = plt.figure(figsize=(16,8))
plt.plot(comp_id, ratios, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
```

```python
#Apply scaling
from sklearn.preprocessing import scale
scaled_X = scale(X, axis=0, with_mean=True, with_std=True, copy=True )
pca = PCA(n_components=400)
X_r = pca.fit(scaled_X).transform(scaled_X)
ratios = pca.explained_variance_ratio_
comp_id = xrange(1,401)             # id number of component
fig = plt.figure(figsize=(16,8))
plt.plot(comp_id, ratios, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
```

### Scaling doesn't seem to make a difference.

## 5. Vary the number k between 1 and <number of datapoints> and plot the Silhouette coefficient as a function of k. What do you observe? What conclusions can you draw?

```python
k_means = xrange(2,len(X))
silhouetteCoefficient = []

for k in k_means:
    km = KMeans(n_clusters=k, init='k-means++', n_init=1 , max_iter = 5, random_state=1)
    km.fit(X)
    labels = km.labels_
    silhouetteCoefficient.append(metrics.silhouette_score(X, labels, metric='euclidean'))

fig = plt.figure(figsize=(8,5))
plt.plot(k_means, silhouetteCoefficient, 'ro-', linewidth=2)
plt.title('Silhouette Coefficient as a function of K')
plt.xlabel('K')
plt.ylabel('Silhouette Coefficient')
```

###Silhoutte coefficients seem to peak around 75 to 125. Let's assume that it peaks at 100. That means that we'll be having 100 different clusters for 40 subjects. So every subject most likely has taken 2 to 3 different pictures.
