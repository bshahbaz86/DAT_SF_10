```python
# imports
from sklearn.datasets import fetch_olivetti_faces
from sklearn.cluster import MiniBatchKMeans
from sklearn.decomposition import PCA
from sklearn.decomposition import RandomizedPCA
from numpy.random import RandomState
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import pairwise_distances
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
import pandas as pd
%matplotlib inline
```

```python
# Load faces data
rng = RandomState(0)
dataset = fetch_olivetti_faces(shuffle=True, random_state=rng)
faces = dataset.data
n_samples, n_features = faces.shape

# global centering
faces_centered = faces - faces.mean(axis=0)

# local centering
faces_centered -= faces_centered.mean(axis=1).reshape(n_samples, -1)
print("Dataset consists of %d faces" % n_samples)
```

## 1. Implement PCA on the dataset, using the sklearn. What do the principal components tell? (is scaling necessary?)

```python
X = faces
y = faces.shape
target_names = faces.shape

# create the model and fit the data
pca = PCA(n_components=10)
X_r = pca.fit(X).transform(X)
ratios = pca.explained_variance_ratio_
print ratios
# print pca.components_
```

```python
comp_id = xrange(1,11)             # id number of component
fig = plt.figure(figsize=(8,5))
plt.plot(comp_id, ratios, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
```

## 2. Change from PCA to RandomizedPCA. Do you obtain the same principal components? Why? Why not?

```python
# create the model and fit the data
rpca = RandomizedPCA(n_components=10, whiten = True)
X_r = rpca.fit(X).transform(X)
rPCAratios = rpca.explained_variance_ratio_
print rPCAratios
# print pca.components_
```

```python
comp_id = xrange(1,11)             # id number of component
fig = plt.figure(figsize=(8,5))
plt.plot(comp_id, rPCAratios, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
```

## 3. Plot the datapoints in the plane of the two principal components. Are any clusters visible?

```python
# create the model and fit the data
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
plt.scatter(X_r[:,0], X_r[:,1])
plt.title('Data Points for TWO Principal Components')
plt.show()
```

## 4. Implement k-means clustering, using sklearn, with k = number of people in the dataset. Does the clustering work? Does each cluster map correctly to one person?

```python
km = KMeans(n_clusters=400,  init='random', n_init=1 , max_iter = 1, random_state=1)
km.fit(X_r)
centroids = km.cluster_centers_
```

```python
def do_kmeans_4(km, data):
    km.fit(data)
    centroids = km.cluster_centers_
#     print "centroids:", centroids
    y = km.predict(data)
    
    
    fig, ax = plt.subplots(1,1, figsize=(8,8))
    for t,marker,c in zip(xrange(4),">ox+","rgby") :
        ax.scatter(data[y == t,0],
                   data[y == t,1],
                   marker=marker,
                   c=c)

    ax.scatter(centroids[:,0],centroids[:,1],marker = 's',c='r')
```

```python
do_kmeans_4(km, X_r)
```

## 5. Vary the number k between 1 and <number of datapoints> and plot the Silhouette coefficient as a function of k. What do you observe? What conclusions can you draw?

```python
km = KMeans(n_clusters=399, random_state=1)
km.fit(X)
labels = km.labels_
metrics.silhouette_score(X, labels, metric='euclidean')
```

```python
for k in xrange(2,5):
    km = KMeans(n_clusters=k, random_state=1)
    km.fit(X)
    labels = km.labels_
    silhouetteCoefficient = metrics.silhouette_score(X, labels, metric='euclidean')

    
    
fig = plt.figure(figsize=(8,5))
plt.plot(k, silhouetteCoefficient, 'ro-', linewidth=2)
plt.title('Silhouette Coefficient as a function of K')
plt.xlabel('K')
plt.ylabel('Silhouette Coefficient')
```
