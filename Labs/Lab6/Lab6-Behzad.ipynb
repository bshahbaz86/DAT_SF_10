##First, import all the modules

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
#import the file
data = pd.read_csv('student_logins.csv')
data.head()
```

##Create two new parameters 1) Duration in minutes 2) Age in minutes

```python
#calculate duration
data.session_start_time = pd.to_datetime(data.session_start_time)
data.session_end_time = pd.to_datetime(data.session_end_time)
data['duration'] = data.session_end_time - data.session_start_time
#convert duration to minutes and append to df
data['minutes']= data['duration'].astype("timedelta64[m]")
#calculate age
data.session_end_time = pd.to_datetime(data.session_end_time)
data.student_account_created = pd.to_datetime(data.student_account_created )
data['age'] = data.session_end_time - data.student_account_created 
#convert duration to minutes and append to df
data['ageInMin']= data['age'].astype("timedelta64[m]")
# calculate number of problems per session duration
data.head()
```

##Get rid of the outliers

```python
#let's get rid of outliers 1) negative durations and 2) very large durations > 600 mins
data = data[data['minutes'] > 0]
data = data[data['minutes'] < 600]
```

##Create a few simple models to assess accuracy of different features

```python
# Number of Problems Completed vs Duration
x = data['problems_completed'].values
y = data['minutes'].values
X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit() 
intercept, slope = results.params

r2 = results.rsquared

plt.plot(x, y, 'bo')
plt.title("Number of Problems Completed vs Duration")
xl = np.array([min(x), max(x)])
yl = intercept + slope * xl
plt.plot(xl, yl, 'r-')
plt.show
```

```python
results.summary()
```

```python
# Total number of logins vs Duration
x = data['student_previous_logins_total'].values
y = data['minutes'].values
X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit() 
intercept, slope = results.params

r2 = results.rsquared

plt.plot(x, y, 'bo')
plt.title("Total number of logins vs Duration")
xl = np.array([min(x), max(x)])
yl = intercept + slope * xl
plt.plot(xl, yl, 'r-')
plt.show
```

```python
results.summary()
```

```python
# Student Age vs Duration of online student
x = data['ageInMin'].values
y = data['minutes'].values
X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit() 
intercept, slope = results.params

r2 = results.rsquared

plt.plot(x, y, 'bo')
plt.title("Student Age vs Duration")
xl = np.array([min(x), max(x)])
yl = intercept + slope * xl
plt.plot(xl, yl, 'r-')
plt.show
```

```python
results.summary()
```

```python
#average mean seem to differ for different class ids so let's categorize different classes based on their mean durations

data.class_id.unique()

avg_duration_e = data.minutes[data.class_id == 'e'].mean()
avg_duration_m = data.minutes[data.class_id == 'm'].mean()
avg_duration_a = data.minutes[data.class_id == 'a'].mean()
avg_duration_g = data.minutes[data.class_id == 'g'].mean()
avg_duration_c = data.minutes[data.class_id == 'c'].mean()


print "Average duration for e students:", avg_duration_e
print "Average duration for m students:", avg_duration_m
print "Average duration for a students:", avg_duration_a
print "Average duration for g students:", avg_duration_g
print "Average duration for c students:", avg_duration_c


#let's create a new feature based on class id. 0 if below 20, 1 for 20 and 2 for over 20
data['Class_label'] = data['class_id']
data.Class_label.replace(['e', 'm', 'a', 'g', 'c'],[1,2,2,1,0],inplace=True)
data.head()
```

```python
#Now, Class Id vs Duration
x = data['Class_label'].values
y = data['minutes'].values
X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit() 
intercept, slope = results.params

r2 = results.rsquared

plt.plot(x, y, 'bo')
plt.title("Class Id vs Duration")
xl = np.array([min(x), max(x)])
yl = intercept + slope * xl
plt.plot(xl, yl, 'r-')
plt.show
```

```python
results.summary()
```

##Final prediction model based on the feature identified above

```python
#Let's combine all of the features and do a final regression
x = data[['problems_completed', 'student_previous_logins_total', 'ageInMin', 'Class_label', 'student_id']].values
y = data['minutes'].values

X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit()
results.summary()
```
